```python
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
import matplotlib.pyplot as plt
import matplotlib.colors as cm
from matplotlib.pyplot import figure
```

```python
n_H = np.load('outputs/sub_grid/n_H.npy')
n_H2 = np.load('outputs/sub_grid/n_H2.npy')
X_H2 = np.load('outputs/sub_grid/X_H2.npy')
n_LW = np.load('outputs/sub_grid/n_LW.npy')
n_LW_ss = np.load('outputs/sub_grid/n_LW_ss.npy')
X_H2_ss = np.load('outputs/sub_grid/X_H2_ss.npy')
n_H2_ss = np.load('outputs/sub_grid/n_H2_ss.npy')
X_CO = np.load('outputs/sub_grid/X_CO.npy')
n_CO = np.load('outputs/sub_grid/n_CO.npy')
pdf = np.load('outputs/sub_grid/pdf.npy')
lambda_jeans = np.load('outputs/sub_grid/lambda_jeans.npy')
```

```python
%%time
plt.figure(figsize=(9,5))
plt.plot(np.log10(n_H), X_H2)
plt.xlabel('log(n_H)')
plt.ylabel('X_H2')
plt.grid(b=True, which='both', axis='both')
plt.title('log(n_H) vs X_H2 - M=10, Z=1solar, G_o=1, n_H_mean=100[H]/cc, T=10K')
plt.savefig('outputs/sub_grid/log(n_H)vsX_H2.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
%%time
plt.figure(figsize=(9,5))
plt.plot(np.log10(n_H), X_CO)
plt.xlabel('log(n_H)')
plt.ylabel('X_CO')
plt.grid(b=True, which='both', axis='both')
plt.title('log(n_H) vs X_CO - M=10, Z=1solar, G_o=1, n_H_mean=100[H]/cc, T=10K')
plt.savefig('outputs/sub_grid/log(n_H)vsX_CO.png', dpi=300, bbox_inches='tight')
plt.show()
```
